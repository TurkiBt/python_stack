```python
import pandas as pa
import numpy as np  
```

### Load data

```python
features = pa.read_csv('features.csv')

print(features)
```

### Shape & Head & Tail

```python
print(features.shape)
print("=============================================================================================")
print(features.head)
print("=============================================================================================")
print(features.tail)
```

### Describe all numerical features and all categorical features

```python
features.describe()  
```

```python
features.dtypes
```

### Mean & Median & Mode & Standard deviation & Sum 

```python
features_num = features.select_dtypes(include=[np.float,np.int64])
print("Mean:",features_num.mean(),sep='\n')
print("================================================")
print("Median:",features_num.median(),sep='\n')
print("================================================")
print("Mode:",features_num.mode(),sep='\n')
print("================================================")
print("Standard deviation:",features_num.std(),sep='\n')
print("================================================")
print("Sum:",features_num.sum(),sep='\n')
```

### Subset your DataFrame based on one condition
### Subset your DataFrame based on two different conditions
### Subset your DataFrame based on three different conditions

```python
features_df1 = features[features['gas']=="air"]
features_df1
```

```python
features_df2 = features[(features["exp"] == 100) & (features["gas"] == "air")]
features_df2
```

```python
features_df3 = features[(features["S3_max"] > 0.0) & (features["gas"] == "air") & (features["S2_max"] > 0.0)]
features_df3
```

### Use loc to slice your DataFrame for the 2 columns that returned the largest sum
### Use iloc to slice your DataFrame for the 2 columns that returned the largest mean

```python
dfSum = features.loc[:,('S2_max', 'S3_max')]
print(dfSum)
```

```python
dfMean = features.iloc[:, 8:10]
print(dfMean)
```
