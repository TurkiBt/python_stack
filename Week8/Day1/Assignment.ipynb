## OpenAQ

```python
import requests

Path_API ='https://u50g7n0cbj.execute-api.us-east-1.amazonaws.com/v2/measurements?date_from=2000-01-01T00%3A00%3A00%2B00%3A00&date_to=2021-05-30T11%3A06%3A00%2B00%3A00&limit=100&page=1&offset=0&sort=desc&radius=1000&country_id=SA&order_by=datetime'

data=requests.get(Path_API)
results=pa.DataFrame(data.json()["results"])

print(results)
print("=================================================================")
print("Cities:",results.groupby("location")["locationId"].count()) # Getting cities are in the data
print("=================================================================")
print("rows and columns",results.shape) # Getting how many rows and columns are in the data

```
